```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

# Add CNN layers
ax.add_patch(patches.Rectangle((1, 8), 2, 1, edgecolor='black', facecolor='lightblue', lw=2))
ax.text(2, 8.5, 'Conv2D (32)', ha='center', va='center', fontsize=10)

ax.add_patch(patches.Rectangle((1, 7), 2, 1, edgecolor='black', facecolor='lightblue', lw=2))
ax.text(2, 7.5, 'BatchNorm', ha='center', va='center', fontsize=10)

ax.add_patch(patches.Rectangle((1, 6), 2, 1, edgecolor='black', facecolor='lightblue', lw=2))
ax.text(2, 6.5, 'Conv2D (64)', ha='center', va='center', fontsize=10)

ax.add_patch(patches.Rectangle((1, 5), 2, 1, edgecolor='black', facecolor='lightblue', lw=2))
ax.text(2, 5.5, 'BatchNorm', ha='center', va='center', fontsize=10)

ax.add_patch(patches.Rectangle((1, 4), 2, 1, edgecolor='black', facecolor='lightgray', lw=2))
ax.text(2, 4.5, 'MaxPool2D', ha='center', va='center', fontsize=10)

# Add BiLSTM
ax.add_patch(patches.Rectangle((4, 6), 2, 2, edgecolor='black', facecolor='lightgreen', lw=2))
ax.text(5, 7, 'BiLSTM', ha='center', va='center', fontsize=10)

# Add Attention mechanism
ax.add_patch(patches.Rectangle((7, 6), 2, 1, edgecolor='black', facecolor='lightcoral', lw=2))
ax.text(8, 6.5, 'Attention', ha='center', va='center', fontsize=10)

# Add Fully Connected layer
ax.add_patch(patches.Rectangle((7, 4), 2, 1, edgecolor='black', facecolor='orange', lw=2))
ax.text(8, 4.5, 'FC Output', ha='center', va='center', fontsize=10)

# Add arrows
arrow_props = dict(facecolor='black', arrowstyle='->', lw=1.5)
ax.annotate('', xy=(3, 8.5), xytext=(4, 7), arrowprops=arrow_props)
ax.annotate('', xy=(6, 7), xytext=(7, 6.5), arrowprops=arrow_props)
ax.annotate('', xy=(8, 6), xytext=(8, 5), arrowprops=arrow_props)

# Display the diagram
plt.show()
```
